```python
#6: Notes

import random
import numpy as np

#STEP 1: Generate the random bit pattern of 1s and 0s of size 𝑚 which is the secret message.
def generate_secret_message():
    x = 1
    y = 0
    m = 10
    secret_message = [] 
    for _ in range(m):
        secret_message.append(random.choice([x,y]))
        
    return secret_message


#STEP 2: Generate a sequence of times when the source will generate the packets. This is based
#inter-packet delay (IPD) distribution of the packets generate by the source. You can intuit
#what is the worst case number of packets that you need.

scale_value = 1

def generate_packet_times(num_packets, scale_value):
    return np.cumsum(np.random.exponential(scale= scale_value , size=num_packets))


#STEP 3, 4, 5: 

def simulation(experiments, encoded_message, buffer_size, initial_buffer_size, scale_value):
    B = buffer_size
    i = initial_buffer_size
    C = i
    next_packet_time = 0
    current_time = 0
    packages_sent_successfully = 0
    packages_sent_unsuccessfully = 0
    overflow_amount = 0
    underflow_amount = 0

    # Step 5:
    while packages_sent_successfully < len(encoded_message):
        next_packet_time = np.random.exponential(scale=scale_value)

        if next_packet_time > current_time:
            arrivals = min(B - C, i)  # Calculate the number of arrivals during the time interval
            C += arrivals

        current_time += next_packet_time

        if C > B:
            overflow_amount += 1
            packages_sent_unsuccessfully += 1
            print("Buffer overflow")
            break
        if C <= 0:
            underflow_amount += 1
            packages_sent_unsuccessfully += 1
            print("Buffer underflow")
            break
        
        packages_sent_successfully += 1

    # Step 5 (continued):
    return packages_sent_successfully, overflow_amount, underflow_amount


# Step 6:
experiments = 100
scale_value = 1
success = 0
success_prob = 0
total_overflow_amount = 0
total_underflow_amount = 0

for _ in range(experiments):
    secret_message = generate_secret_message()
    buffer_size = len(secret_message)
    initial_buffer_size = len(secret_message)
    current_buffer_size = initial_buffer_size
    packet_times = generate_packet_times(len(secret_message), scale_value)

    successful_packages, overflow_amount, underflow_amount = simulation(
        experiments, secret_message, buffer_size, initial_buffer_size, scale_value
    )

    if successful_packages == len(secret_message):
        success += 1

    total_overflow_amount += overflow_amount
    total_underflow_amount += underflow_amount

# Calculate probabilities
success_prob = success / experiments
overflow_prob = total_overflow_amount / experiments
underflow_prob = total_underflow_amount / experiments

print("Successful packet delivery probability: ", success_prob)
print("Overflow Prob: ", overflow_prob)
print("Underflow Prob: ", underflow_prob)

```
